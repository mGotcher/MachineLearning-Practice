```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

# Example 1 - A simple cost function
### $$f(x) = x^2 + x + 1$$

```python
def f(x): 
    # * is multiply ** is to the power of
    return x**2 + x + 1
```

### Make data


```python
#Creates data that creates a sequence of numbers between -3 and 3 and gives us 100 numbers in between
x_1 = np.linspace(start=-3, stop =3, num=500) 

#If you want to see it
#print(x_1)
```

```python
#Plot function
plt.xlim([-3, 3])
plt.ylim(0,13)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.plot(x_1, f(x_1))
plt.show()
```

## Slope & Derivitives

```python
def df(x):
    return (2 * x) + 1
```

```python
#Plot function and derivitave together
plt.figure(figsize=[15,5])
plt.subplot(1, 2, 1)
plt.xlim([-3, 3])
plt.ylim(0,13)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.title('Cost function', fontsize=18)
plt.plot(x_1, f(x_1), color='green', linewidth=3)
plt.subplot(1, 2, 2)
plt.xlim([-3, 3])
plt.ylim(-6,8)
plt.title('Deriviative of Cost function', fontsize=18)
plt.grid()
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)

plt.plot(x_1, df(x_1), color='red', linewidth=3)

plt.show()
```

# Gradient Descent

```python
#Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.0001

x_list = [new_x]
slope_list = [df(new_x)]
for n in range(500):
    previous_x = new_x
    gradient=df(previous_x)
    new_x = previous_x - (step_multiplier * gradient)
    
    step_size = abs(new_x - previous_x)
    
    x_list.append(new_x)
    slope_list.append(df(new_x))
    if step_size < precision:
        break
    #If you want to see what this looks like
    #print("\n\nprevious_x: ", previous_x)
    #print("gradient: ", gradient)
    #print("new_x: ", new_x)
print('Local Minimum occurs at: ', new_x)
print('Slope or d(x) looks like: ', df(new_x))
print('f(x) value or cost at this pouint is: ', f(new_x))
```

```python
#Superimpose the gradient descent calculations on plot
plt.figure(figsize=[15,5])
plt.subplot(1, 2, 1)
plt.xlim([-3, 3])
plt.ylim(0,13)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.title('Cost function', fontsize=18)
plt.plot(x_1, f(x_1), color='green', linewidth=3, alpha=0.6)
values = np.array(x_list)
plt.scatter(x_list, f(values), color = 'blue', s=100, alpha=0.5)
plt.subplot(1, 2, 2)
plt.xlim([-3, 3])
plt.ylim(-6,8)
plt.title('Deriviative of Cost function', fontsize=18)
plt.grid()
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)

plt.plot(x_1, df(x_1), color='red', linewidth=3, alpha =0.6)
plt.scatter(x_list, df(values), color = 'blue', s=100, alpha=0.5)

plt.show()
```
